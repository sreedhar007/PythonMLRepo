```
def myreduce(func,sequence): 
  #Taking First element in sequence for intial calculation
  result=sequence[0]
  #Iterating all elements in sequence
  for item in sequence[1:]:
    #Calling required function and calculating result
    result=func(result,item)
  print('result after applying reduce function is:',result)



def sum(a,b):
  return a+b

def mul(a,b):
  return a*b



l=[1,2,3,4]
myreduce(sum,l)
myreduce(mul,l)
```
