```
class Parent:
  def sideValues():
    a=int(input('Enter a side'))
    b=int(input('Enter b side'))
    c=int(input('Enter b side'))
    return a,b,c

class Child(Parent):
  def area(self):
    a,b,c=Parent.sideValues()
    s=(a+b+c)/2
    area= (s*(s-a)*(s-b)*(s-c)) ** 0.5 
    print('area of triangle is: ',area)

areaT=Child()
areaT.area()
```
