```
def myfilter(func,sequence):
  #Creating empty list to store result
  result=[]
  #Iterating all elements in sequence
  for item in sequence:
    if(func(item)):
      result.append(item)
    else:
      continue
  print('result after filtering is:',result)

def isVowel(variable): 
    letters = ['a', 'e', 'i', 'o', 'u'] 
    if (variable in letters): 
        return True
    else: 
        return False
myfilter(isVowel,['q','a','r','u','i','pp','oo'])
```
