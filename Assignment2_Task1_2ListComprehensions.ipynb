```
inputlist=['x','y','z']
result=[var*num for var in inputlist for num in range(1,5)]
print("['x','y','z'] is transformed into:",result)
```

```
inputlist=['x','y','z']

result=[var*num for num in range(1,5) for var in inputlist]
print("['x','y','z'] is transformed into:",result)
```

```
input=[2,3,4]
result=[num+digit for num in input for digit in range(0,3)]
print("['2','3','4'] is transformed into:",result)
```

```
input=[2,3,4,5]
result=[[item+digit for item in input] for digit in range(0,3)]
print("['2','3','4','5'] is transformed into:",result)
```

```
input_list=[1,2,3]
result = [ (b,a) for a in input_list for b in input_list]
print("[1,2,3] is transformed into:",  result)
```
